```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cmap='viridis', cbar=False)
plt.title('Missing Values')
plt.show()
```

Heatmap indicates missing values for columns 'age', 'sex', and 'smoking'

```python
df.hist(figsize=(16, 14))
plt.suptitle('Histograms of Columns')
plt.show()
```

The sample featured 300 subjects comprising of both males (n=200) and females (n=100) aged 40-95.

One third of the sample were smokers and had high blood pressure. No significant correlation was found between the two variables.


```python
numerical_columns = df.select_dtypes(include=['int','float'])

sns.pairplot(numerical_columns)
plt.title('Pairplot of Numerical Features')
plt.show()
```

The pairplot showed no significant correlations between any of the continuous variables.

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Moderate positive correlations were found between 'smoking' and 'sex' (r=0.45) and between 'serum_creatinine' and 'death_event' (r=0.29). Majority of subjects who had smoked were male and higher levels of serum creatinine correlated with death event.

A moderate negative correlation was found between 'time' and 'death_event' (r=-0.53) indicating a decrease in the number of deaths as the number of follow-up days increased.

A weak negative correlation was found between 'ejection_fraction' and 'death_event' (r=-0.27) suggesting subjects with a lower percentage of blood leaving the heart at each contraction (14-38%) correlated with death event.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.boxplot(df['platelets'])

# Add title and labels
plt.title('Boxplot of Platelet Column')
plt.ylabel('Platelet Values')

# Show the plot
plt.show()
```

The 'platelets' variable produced a boxplot with 50% of the data ranging between the 25th percentile (212000 kiloplatelets/mL) and the 75th percentile (304000 kiloplatelets/mL) with a median of 262000 kiloplatelets/mL.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

non_binary_cols = df.loc[:, df.nunique() > 2]

# Create boxplots for non-binary columns
non_binary_cols.boxplot()

# Set title and labels
plt.title('Boxplot of Non-Binary Columns')
plt.ylabel('Values')

plt.xticks(rotation=45)
plt.show()
```

The only other variable that produced an observable boxplot was 'creatinine_phosphokinase(CPK)' which had an interquartile range (IQR) of 115-582 mcg/L with a median of 250 mcg/L.

Features with the most outliers included 'CPK' and 'serum_creatine' (29 each) and 'platelets' (21)

```python
binary_columns = [col for col in df.columns if df[col].nunique() == 2]

# Plot bar charts for each binary column
for col in binary_columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col)
    plt.title(f'Bar Chart for {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

The number of participants who were male was double the number of females. The same size discrepencey was found in both the 'smoking' and 'high blood pressure' columns with 2/3 subjects being non-smokers and 1/3 of subjects having high blood pressure.

```python
df_dropped = df.dropna()

df_imputed = df.fillna(df.mean())

from sklearn.impute import KNNImputer

imputer = KNNImputer()

df_knn_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

print("Shape of original dataset:", df.shape)
print("Shape of dataset after dropping missing values:", df_dropped.shape)
print("Shape of dataset after imputation with mean:", df_imputed.shape)
print("Shape of dataset after KNN imputation:", df_knn_imputed.shape)
```

```python
cleaned_data_mean = df.fillna(df.mean)
cleaned_data_mean
```

The above fillna(mean) cleaning method was used to fill missing values with the mean value of each column.

```python
cleaned_data_median = df.fillna(df.median)
cleaned_data_median
```

The above fillna(median) cleaning method was used to fill missing values with the median value of each column.

```python
cleaned_data_mode = df.fillna(df.mode().iloc[0])
cleaned_data_mode
```

The above fillna(mode) cleaning method was used to fill missing values with the mode (most frequent value) of each column.

```python
cleaned_data_filled = df.fillna(method='bfill')
cleaned_data_filled
```

The above backfill method of cleaning was used to fill missing values with the values of the next valid observation in the DataFrame.
